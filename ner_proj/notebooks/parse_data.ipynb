## Utility for parsing the input files 

```python
def translate_iob(lable):
    return lable

def create_df1(filename, small_sample = False):
    df_freq = pd.DataFrame(columns=('token', 'pos', 'chunk', 'lable'))
    df_features = pd.DataFrame(columns=('features', 'lables'))
    
    #TRAIN_DATA = [
    #("Who is Shaka Khan?", {"entities": [(7, 17, "PERSON")]}),
    #("I like London and Berlin.", {"entities": [(7, 13, "LOC"), (18, 24, "LOC")]}),
    #]
    df_spacy = pd.DataFrame(columns=('text', 'entities')) # text and set of entriies
    
    try:
        f = open(filename, "r")
    except: 
        print('Unable to open an input file ' + filename)
    else:
        j = 0
        list_features = []
        list_lables = []
        list_of_ent_tuples = []
        prev = -1
        
        try:
            f.readline()
        except:
            print('error reading from input file' + filename)
        else:    
            for i, x in enumerate(f):
                vals = x.split()
                if len(vals) > 0:
                    # record entries per individual token, in format: 'token', 'pos', 'chunk', 'lable'
                    df_freq.loc[i-1] = vals
                    token = vals[0]
                    list_features.append(token)
                    lable = vals[3]
                    list_lables.append(lable)
                    t_len = len(token)
                    if (lable != 'O'): 
                        list_of_ent_tuples.append((prev+1, prev+1+t_len, translate_iob(lable)))
                    prev = prev + t_len + 1
                else:
                    if len(list_features) > 1 or len(list_features) == 1 and list_features[0] != "-DOCSTART-":
                        # record samples, each sample is a sentence, format: [tokens], [lables]
                        df_features.loc[j] = [list_features, list_lables]
                        str = ' '.join(list_features)
                        df_spacy.loc[j] = [str, list_of_ent_tuples]

                        j = j + 1
                        if(j % 50 == 0): 
                            print('progres: formatted ', j ,'sentences in ', filename)
                            if (small_sample == True) and (j > 51): 
                                return  df_freq, df_features, df_spacy
                    list_features = []
                    list_lables = []
                    list_of_ent_tuples = []
                    prev = -1
        finally:
            f.close()
    return df_freq, df_features, df_spacy
```
